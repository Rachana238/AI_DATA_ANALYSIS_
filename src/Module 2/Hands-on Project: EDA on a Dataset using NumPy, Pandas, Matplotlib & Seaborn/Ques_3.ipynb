```python
import pandas as pd
import seaborn as sns

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Calculate the ratio of petal length to petal width
iris['petal_ratio'] = iris['petal_length'] / iris['petal_width']

# Display the first few rows with the new column
print(iris.head())

# You can also display some descriptive statistics of the new column
print("\nDescriptive statistics of the petal ratio:")
print(iris['petal_ratio'].describe())
```
