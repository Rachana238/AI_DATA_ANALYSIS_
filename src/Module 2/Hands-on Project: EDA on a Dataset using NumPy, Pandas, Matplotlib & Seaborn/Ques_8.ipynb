```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
data = {
    'A': [1, 2, None, 4, 5],
    'B': [5, None, None, 8, 10],
    'C': [None, 2, 3, 4, 5]
}

df = pd.DataFrame(data)

# Impute missing values with the median (as in the previous example)
df_imputed = df.fillna(df.median(numeric_only=True))

# Create the pair plot
sns.pairplot(df_imputed)
plt.suptitle("Pairwise Relationships", y=1.02) # Add a title, adjust vertical position
plt.show()

# Title: Advanced Data Visualization
# Description: Create a pair plot for the dataset to display pairwise relationships.
```
