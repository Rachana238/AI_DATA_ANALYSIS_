```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

def plot_petal_length_distribution():
    """Loads the Iris dataset and plots the distribution of petal length
        with error handling and memory management.
    """
    try:
        iris = sns.load_dataset('iris')
    except Exception as e:
        raise Exception(f"Error loading the dataset: {e}")

    if not isinstance(iris, pd.DataFrame):
        raise TypeError("Loaded data is not a Pandas DataFrame.")

    if 'petal_length' not in iris.columns:
        raise ValueError("'petal_length' not found in the dataset columns.")

    sns.histplot(iris['petal_length'], kde=True)
    plt.title('Distribution of Petal Length')
    plt.xlabel('Petal Length (cm)')
    plt.ylabel('Frequency')
    plt.show()
    plt.close() # Close the plot to free up memory

# --- Main execution ---
if __name__ == "__main__":
    try:
        plot_petal_length_distribution()
        print("Petal length distribution plotted successfully.")
    except (ValueError, TypeError, Exception) as e:
        print(f"An error occurred: {e}")
```
