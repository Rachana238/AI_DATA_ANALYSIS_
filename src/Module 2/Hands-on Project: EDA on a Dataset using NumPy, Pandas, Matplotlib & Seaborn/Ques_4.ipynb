```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

def plot_petal_length_distribution(data_source='iris'):
    """Loads a dataset and plots the distribution of petal length.

    Args:
        data_source (str, optional): The name of the dataset to load
                                      from seaborn. Defaults to 'iris'.

    Raises:
        ValueError: If 'petal_length' column is not found in the dataset.
        TypeError: If the loaded data is not a Pandas DataFrame.
        Exception: If any other error occurs during data loading.
    """
    try:
        data = sns.load_dataset(data_source)
    except Exception as e:
        raise Exception(f"Error loading the dataset '{data_source}': {e}")

    if not isinstance(data, pd.DataFrame):
        raise TypeError("Loaded data is not a Pandas DataFrame.")

    if 'petal_length' not in data.columns:
        raise ValueError(f"'petal_length' not found in the '{data_source}' dataset columns.")

    plt.figure(figsize=(8, 6))  # Set figure size for better visualization
    sns.histplot(data['petal_length'], kde=True)
    plt.title(f'Distribution of Petal Length in {data_source.capitalize()} Dataset')
    plt.xlabel('Petal Length (cm)')
    plt.ylabel('Frequency')
    plt.tight_layout()  # Adjust layout to prevent overlapping elements
    plt.show()
    plt.close()

if __name__ == "__main__":
    try:
        plot_petal_length_distribution()  # Using default 'iris' dataset
        # You can also try: plot_petal_length_distribution('tips')
        print("Petal length distribution plotted successfully.")
    except (ValueError, TypeError, Exception) as e:
        print(f"An error occurred: {e}")
```
