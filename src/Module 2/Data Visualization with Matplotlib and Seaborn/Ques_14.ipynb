```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Customize the line plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, color='skyblue', linestyle='--', linewidth=2, marker='o', markersize=8, markerfacecolor='lightcoral', markeredgecolor='salmon')

# Add title and labels with styling
plt.title('Sine Wave with Customized Style', fontsize=16, fontweight='bold')
plt.xlabel('X-axis', fontsize=12, style='italic')
plt.ylabel('Y-axis (sin(x))', fontsize=12, style='italic')

# Add grid with custom style
plt.grid(True, linestyle=':', alpha=0.7)

# Add a legend (optional, but good practice)
plt.legend(['Sine Wave'], loc='upper right')

# Customize the appearance of ticks
plt.xticks(fontsize=10, rotation=45)
plt.yticks(fontsize=10)

# Add annotations (optional example)
plt.annotate('Peak', xy=(np.pi/2, 1), xytext=(np.pi/2 + 1, 1.2),
             arrowprops=dict(facecolor='black', shrink=0.05),
                          fontsize=10)

                          # Show the plot
                          plt.tight_layout() # Adjust layout to prevent labels from overlapping
                          plt.show()
```
