```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Customize the line plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, color='blue', linestyle='--', marker='o')

# Add a title and labels
plt.title('Styled Sine Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Add a grid
plt.grid(True)

# Show the plot
plt.show()
```
