```python
import matplotlib.pyplot as plt
import numpy as np

# Product names
products = ['A', 'B', 'C', 'D', 'E']
# Sales values (in thousands)
sales = np.array([23, 75, 57, 90, 65])

# Create the bar plot
plt.figure(figsize=(8, 6))
plt.bar(products, sales, color='skyblue')

# Add labels and title
plt.xlabel('Products', fontname='sans-serif', fontsize=12)
plt.ylabel('Sales (in thousands)', fontname='sans-serif', fontsize=12)
plt.title('Number of Sales for Each Product', fontname='sans-serif', fontsize=14, fontweight='bold')

# Add grid
plt.grid(axis='y', linestyle='--')

# Customize tick labels (optional)
plt.xticks(fontname='sans-serif', fontsize=10)
plt.yticks(fontname='sans-serif', fontsize=10)

# Show the plot
plt.tight_layout()
plt.show()
```
