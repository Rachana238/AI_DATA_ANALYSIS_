```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
np.random.seed(42)
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.random.rand(50)
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 10, size=5)

# Create a figure and a set of subplots (2 rows, 2 columns)
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))
fig.suptitle('Combined Plots', fontsize=16)

# Plot 1: Line plot
axes[0, 0].plot(x, y1, label='sin(x)', color='blue')
axes[0, 0].set_xlabel('X-axis')
axes[0, 0].set_ylabel('sin(x)')
axes[0, 0].set_title('Sine Wave')
axes[0, 0].legend()
axes[0, 0].grid(True)

# Plot 2: Line plot
axes[0, 1].plot(x, y2, label='cos(x)', color='green')
axes[0, 1].set_xlabel('X-axis')
axes[0, 1].set_ylabel('cos(x)')
axes[0, 1].set_title('Cosine Wave')
axes[0, 1].legend()
axes[0, 1].grid(True)

# Plot 3: Scatter plot
axes[1, 0].scatter(x[:50], y3, label='Random Data', color='red', marker='o')
axes[1, 0].set_xlabel('X-axis')
axes[1, 0].set_ylabel('Random Value')
axes[1, 0].set_title('Scatter Plot')
axes[1, 0].legend()
axes[1, 0].grid(True)

# Plot 4: Bar chart
axes[1, 1].bar(categories, values, color='purple')
axes[1, 1].set_xlabel('Category')
axes[1, 1].set_ylabel('Value')
axes[1, 1].set_title('Bar Chart')
axes[1, 1].grid(axis='y')

# Adjust layout to prevent overlapping titles/labels
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust for suptitle

# In Colab, plots are displayed automatically when you run the cell
# You don't need plt.show() explicitly.
```
