```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for your plots (replace with your actual data)
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = x
y4 = x**2

# Create a figure and a set of subplots
# The numbers in plt.subplots() indicate the number of rows and columns of subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8)) # Adjust figsize as needed

# Plot data on each subplot
axes[0, 0].plot(x, y1, label='sin(x)')
axes[0, 0].set_title('Sine Wave')
axes[0, 0].set_xlabel('X-axis')
axes[0, 0].set_ylabel('Y-axis')
axes[0, 0].legend()

axes[0, 1].plot(x, y2, label='cos(x)', color='orange')
axes[0, 1].set_title('Cosine Wave')
axes[0, 1].set_xlabel('X-axis')
axes[0, 1].set_ylabel('Y-axis')
axes[0, 1].legend()

axes[1, 0].plot(x, y3, label='y = x', color='green')
axes[1, 0].set_title('Linear Plot')
axes[1, 0].set_xlabel('X-axis')
axes[1, 0].set_ylabel('Y-axis')
axes[1, 0].legend()

axes[1, 1].plot(x, y4, label='y = x^2', color='red')
axes[1, 1].set_title('Quadratic Plot')
axes[1, 1].set_xlabel('X-axis')
axes[1, 1].set_ylabel('Y-axis')
axes[1, 1].legend()

# Adjust layout to prevent overlapping titles and labels
plt.tight_layout()

# Display the figure
plt.show()
```
