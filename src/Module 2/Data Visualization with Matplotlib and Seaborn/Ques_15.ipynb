```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Attempt to load only the required features of the iris dataset with error handling
try:
    iris = sns.load_dataset('iris')[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
    print("Required features of the iris dataset loaded successfully.") # This line was incorrectly indented
except Exception as e:
    print(f"Error loading the iris dataset: {e}")
    exit()

# Check for missing values in the relevant columns
if iris.isnull().any().any():
    print("Warning: Missing values found in the dataset. Consider handling them before calculating correlation.")

# Calculate the correlation matrix
correlation_matrix = iris.corr()

# Basic styled heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Iris Feature Correlations')
plt.show()
```
