```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the iris dataset
iris = sns.load_dataset('iris')

# Calculate the correlation matrix
correlation_matrix = iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].corr()

# Advanced styling for the heatmap
plt.figure(figsize=(8, 7))
sns.heatmap(correlation_matrix,
            annot=True,             # Display correlation values in the heatmap
            cmap='coolwarm',        # Use a diverging colormap
            fmt=".2f",              # Format the annotation values to two decimal places
            linewidths=.5,         # Add lines between cells
            cbar_kws={'shrink': .8}, # Adjust the size of the color bar
            annot_kws={'size': 10, 'weight': 'bold'}) # Style the annotation text

# The following lines were incorrectly indented
plt.title('Correlation Matrix of Iris Features', fontsize=16, fontweight='bold')
plt.xticks(fontsize=10)
plt.yticks(fontsize=10, rotation=0) # Ensure y-axis labels are horizontal
plt.tight_layout()
plt.show()
```
