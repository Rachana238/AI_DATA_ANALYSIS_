```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the histogram
marks = np.array([23, 45, 56, 78, 89, 90, 67, 88, 76, 59])
num_bins = 10
hist_color = 'lightcoral'
edge_color = 'black'

# Create the figure and axes object
fig, ax = plt.subplots(figsize=(8, 6))

# Create the histogram
n, bins, patches = ax.hist(marks, bins=num_bins, color=hist_color, edgecolor=edge_color)

# Add labels and title
ax.set_xlabel('Marks')
ax.set_ylabel('Frequency')
ax.set_title('Distribution of Marks Obtained by Students')
ax.grid(axis='y', linestyle='--')

# Ensure a tight layout
fig.tight_layout()

# Show the plot
plt.show()
```
