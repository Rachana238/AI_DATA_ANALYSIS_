```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Data from Question 1 (Product Sales)
products = ['A', 'B', 'C', 'D', 'E']
sales = np.array([23, 75, 57, 90, 65])

# Data from Question 2 (Height vs. Weight)
heights = [160, 170, 165, 180, 175]
weights = [55, 65, 60, 80, 76]

# Create subplots (1 row, 2 columns)
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot 1: Bar plot for product sales
axes[0].bar(products, sales, color='skyblue')
axes[0].set_xlabel('Products')
axes[0].set_ylabel('Sales (in thousands)')
axes[0].set_title('Product Sales')

# Plot 2: Scatter plot for height vs. weight
axes[1].scatter(heights, weights, color='salmon')
axes[1].set_xlabel('Height (cm)')
axes[1].set_ylabel('Weight (kg)')
axes[1].set_title('Height vs. Weight')

# Adjust layout to prevent overlapping titles/labels
plt.tight_layout()

# Display the plots
plt.show()
```
