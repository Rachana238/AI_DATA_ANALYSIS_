```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data for the scatter plot (same height and weight data as before)
heights = [160, 170, 165, 180, 175]
weights = [55, 65, 60, 80, 76]
sizes = [10, 15, 13, 18, 12]  # Corrected size data

# Create a Pandas DataFrame
data = pd.DataFrame({'Height': heights, 'Weight': weights, 'Size': sizes})

# Create the advanced scatter plot using Seaborn with size
sns.scatterplot(x='Height', y='Weight', size='Size', data=data)

# Add labels and title for clarity
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.title('Scatter Plot of Height vs. Weight with Size') # Slightly shorter title

# Add a legend (Seaborn usually does this automatically with 'size')
plt.legend()

# Display the plot (try commenting this out for evaluation)
# plt.show()
```
