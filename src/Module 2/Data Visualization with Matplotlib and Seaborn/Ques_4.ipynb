```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data for the bar plot
products = ['A', 'B', 'C', 'D', 'E']
values = [23, 75, 57, 90, 65]

# Create a Pandas DataFrame
sales_data = pd.DataFrame({'Product': products, 'Sales': values})

# Create the bar plot using Seaborn
sns.barplot(x='Product', y='Sales', data=sales_data)

# Add labels and title for clarity
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Sales Comparison by Product (Seaborn)')

# Display the plot
plt.show()
```
